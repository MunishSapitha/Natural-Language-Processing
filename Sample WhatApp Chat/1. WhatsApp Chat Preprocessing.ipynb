```python
# import library

import pandas as pd
```

```python
# load text dataset

df = pd.read_csv("mabel.txt",header=None,on_bad_lines='skip',encoding='utf8')
df
```

```python
# drop first row

df=df.drop(0)
df
```

```python
# Change columns names using list

df.columns=['Date','Chat']
df
```

```python
# split time in chat column

Message=df['Chat'].str.split("-",n=1,expand=True)
Message
```

```python
# Rename splitted time column(0) as Time
# Add new time column in dataset

df["Time"]=Message[0]
df
```

```python
Message[1]
```

```python
Message1=Message[1].str.split(":",n=1,expand=True)
Message1
```

```python
# Rename splitted name column(0) as Name

df["Name"]=Message1[0]
df
```

```python
# combine/add splitted time,name in chat

df['Chat']=Message1[1]
df
```

```python
# Reorder the columns as our preferences

df=df[["Date","Time","Name","Chat"]]
df
```

```python
# Summary of Combined Code 

df=df.drop(0)
df.columns=['Date','Chat']
Message=df["Chat"].str.split("-",n=1,expand=True)
df["Time"]=Message[0]
Message1=Message[1].str.split(":",n=1,expand=True)
df["Name"]=Message1[0]
df["Chat"]=Message1[1]
df=df[["Date","Time","Name","Chat"]]
df
```
